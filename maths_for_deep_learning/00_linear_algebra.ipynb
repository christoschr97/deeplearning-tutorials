```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vector
v = np.array([3, 4])

# Plot setup
plt.figure()
plt.axhline(0, color='gray', linewidth=1)
plt.axvline(0, color='gray', linewidth=1)

# Draw the vector as an arrow
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue')

# Labels and limits
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title(f"Vector v = {v}")
plt.show()
```

```python
import numpy as np

def magnitude(vec):
    return np.linalg.norm(vec)

def angle_between(v1, v2):
    dot = np.dot(v1, v2)
    return np.degrees(np.arccos(dot / (np.linalg.norm(v1) * np.linalg.norm(v2))))

v = np.array([3, 4])
u = np.array([1, 2])
sum_vec = v + u

print("Magnitude of v:", magnitude(v))
print("Magnitude of u:", magnitude(u))
print("Magnitude of v+u:", magnitude(sum_vec))

print("Angle between v and (v+u):", angle_between(v, sum_vec))
print("Angle between u and (v+u):", angle_between(u, sum_vec))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors by their start and end coordinates
# Vector AB: A(0,4) to B(1,1)
A = np.array([0, 4])
B = np.array([1, 1])
AB = B - A  # Direction vector

# Vector GD: G(1,0) to D(3,2)
G = np.array([1, 0])
D = np.array([3, 2])
GD = D - G  # Direction vector

# Vector EZ: E(4,0) to Z(3,3)
E = np.array([4, 0])
Z = np.array([3, 3])
EZ = Z - E  # Direction vector

# Create the plot
plt.figure(figsize=(10, 8))
plt.axhline(0, color='gray', linewidth=0.5, alpha=0.7)
plt.axvline(0, color='gray', linewidth=0.5, alpha=0.7)

# Plot Vector AB (from A to B)
plt.quiver(A[0], A[1], AB[0], AB[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.005, label=f'AB: A{tuple(A)}  B{tuple(B)}')

# Plot Vector GD (from G to D)
plt.quiver(G[0], G[1], GD[0], GD[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.005, label=f'GD: G{tuple(G)}  D{tuple(D)}')

# Plot Vector EZ (from E to Z)
plt.quiver(E[0], E[1], EZ[0], EZ[1], angles='xy', scale_units='xy', scale=1, 
           color='green', width=0.005, label=f'EZ: E{tuple(E)}  Z{tuple(Z)}')

# Mark the points
points = [A, B, G, D, E, Z]
labels = ['A', 'B', 'G', 'D', 'E', 'Z']
for point, label in zip(points, labels):
    plt.plot(point[0], point[1], 'ko', markersize=6)
    plt.annotate(label, (point[0], point[1]), xytext=(5, 5), textcoords='offset points')

# Set up the plot
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.legend()
plt.title('Vector Visualization: AB, GD, and EZ')
plt.axis('equal')
plt.show()

# Print vector information
print("Vector Information:")
print(f"Vector AB = {AB} (magnitude: {np.linalg.norm(AB):.2f})")
print(f"Vector GD = {GD} (magnitude: {np.linalg.norm(GD):.2f})")
print(f"Vector EZ = {EZ} (magnitude: {np.linalg.norm(EZ):.2f})")
```

```python
# Vector addition visualizations showing vectors in their original positions
import numpy as np
import matplotlib.pyplot as plt

# Calculate vector sums
AB_plus_GD = AB + GD
AB_plus_GD_plus_EZ = AB + GD + EZ

# Create subplot for both visualizations
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Plot 1: AB + GD using tip-to-tail method
ax1.axhline(0, color='gray', linewidth=0.5, alpha=0.7)
ax1.axvline(0, color='gray', linewidth=0.5, alpha=0.7)

# Plot original vectors in their positions
ax1.quiver(A[0], A[1], AB[0], AB[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.005, label=f'AB from A{tuple(A)}')
ax1.quiver(G[0], G[1], GD[0], GD[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.005, label=f'GD from G{tuple(G)}')

# For addition, show GD starting from the tip of AB (tip-to-tail)
ax1.quiver(B[0], B[1], GD[0], GD[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.003, alpha=0.5, label='GD moved to tip of AB')

# Show the resultant vector AB + GD from A to the final point
final_point_1 = A + AB + GD
ax1.quiver(A[0], A[1], AB_plus_GD[0], AB_plus_GD[1], angles='xy', scale_units='xy', scale=1, 
           color='purple', width=0.008, label=f'AB + GD = {AB_plus_GD}')

# Mark important points
points_1 = [A, B, G, D, final_point_1]
labels_1 = ['A', 'B', 'G', 'D', 'A+AB+GD']
for point, label in zip(points_1, labels_1):
    ax1.plot(point[0], point[1], 'ko', markersize=4)
    ax1.annotate(label, (point[0], point[1]), xytext=(3, 3), textcoords='offset points', fontsize=8)

ax1.set_xlim(-1, 6)
ax1.set_ylim(-1, 5)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True, alpha=0.3)
ax1.legend(fontsize=8)
ax1.set_title('Vector Addition: AB + GD (Tip-to-Tail)')
ax1.set_aspect('equal')

# Plot 2: AB + GD + EZ using tip-to-tail method
ax2.axhline(0, color='gray', linewidth=0.5, alpha=0.7)
ax2.axvline(0, color='gray', linewidth=0.5, alpha=0.7)

# Plot original vectors in their positions
ax2.quiver(A[0], A[1], AB[0], AB[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.005, label=f'AB from A{tuple(A)}')
ax2.quiver(G[0], G[1], GD[0], GD[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.005, label=f'GD from G{tuple(G)}')
ax2.quiver(E[0], E[1], EZ[0], EZ[1], angles='xy', scale_units='xy', scale=1, 
           color='green', width=0.005, label=f'EZ from E{tuple(E)}')

# For addition, show tip-to-tail chain: AB -> GD -> EZ
# GD starting from tip of AB
ax2.quiver(B[0], B[1], GD[0], GD[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.003, alpha=0.5)
# EZ starting from tip of (AB + GD)
intermediate_point = B + GD
ax2.quiver(intermediate_point[0], intermediate_point[1], EZ[0], EZ[1], angles='xy', scale_units='xy', scale=1, 
           color='green', width=0.003, alpha=0.5)

# Show the final resultant vector AB + GD + EZ
final_point_2 = A + AB + GD + EZ
ax2.quiver(A[0], A[1], AB_plus_GD_plus_EZ[0], AB_plus_GD_plus_EZ[1], angles='xy', scale_units='xy', scale=1, 
           color='orange', width=0.008, label=f'AB + GD + EZ = {AB_plus_GD_plus_EZ}')

# Mark important points
points_2 = [A, B, G, D, E, Z, intermediate_point, final_point_2]
labels_2 = ['A', 'B', 'G', 'D', 'E', 'Z', 'B+GD', 'Final']
for point, label in zip(points_2, labels_2):
    ax2.plot(point[0], point[1], 'ko', markersize=4)
    ax2.annotate(label, (point[0], point[1]), xytext=(3, 3), textcoords='offset points', fontsize=8)

ax2.set_xlim(-1, 6)
ax2.set_ylim(-1, 5)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.grid(True, alpha=0.3)
ax2.legend(fontsize=8)
ax2.set_title('Vector Addition: AB + GD + EZ (Tip-to-Tail)')
ax2.set_aspect('equal')

plt.tight_layout()
plt.show()

# Print results
print("Vector Addition Results:")
print(f"AB = {AB} (from A{tuple(A)} to B{tuple(B)})")
print(f"GD = {GD} (from G{tuple(G)} to D{tuple(D)})")
print(f"EZ = {EZ} (from E{tuple(E)} to Z{tuple(Z)})")
print(f"\nAB + GD = {AB_plus_GD} (magnitude: {np.linalg.norm(AB_plus_GD):.2f})")
print(f"AB + GD + EZ = {AB_plus_GD_plus_EZ} (magnitude: {np.linalg.norm(AB_plus_GD_plus_EZ):.2f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Points
A = np.array([0, 4])
B = np.array([1, 1])
G = np.array([1, 0])
D = np.array([3, 2])

# Displacement vectors
AB = B - A          # (1, -3)
GD = D - G          # (2,  2)
R  = AB + GD        # (3, -1)
Z  = A + R          # endpoint of the sum if we start at A

fig, ax = plt.subplots(figsize=(7, 6))

# Plot points
for p, name in [(A, "A"), (B, "B"), (G, "G"), (D, "D"), (Z, "Z")]:
    ax.scatter(*p)
    ax.text(p[0] + 0.05, p[1] + 0.05, name)

# Draw AB from A
ax.arrow(A[0], A[1], AB[0], AB[1],
         length_includes_head=True, head_width=0.08, head_length=0.12,
         label="AB")

# Draw GD translated to start at B (tail-to-head with AB)
ax.arrow(B[0], B[1], GD[0], GD[1],
         length_includes_head=True, head_width=0.08, head_length=0.12,
         label="GD (translated)")

# Draw the resultant R = AB + GD from A to Z
ax.arrow(A[0], A[1], R[0], R[1],
         length_includes_head=True, head_width=0.08, head_length=0.12,
         linestyle="--", label="AB + GD")

# Formatting
ax.set_title("Vector Addition with Tail-to-Head Translation")
ax.set_xlabel("x"); ax.set_ylabel("y")
ax.set_aspect("equal", adjustable="box")
ax.grid(True)
ax.legend()
ax.set_xlim(-0.5, 4.5); ax.set_ylim(-0.5, 4.5)

plt.show()

```
